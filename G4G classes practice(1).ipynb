```python
#Lambda activity
class Solution:
    def binarysearch(self, arr, k):
        index = 0
        check = lambda x : index if x == k else -1
        for n in arr:
            if check(n) != -1:
                answer = index
                break
            else:
                answer = -1
            index += 1
        return(answer)
```
