```python
class Solution:
    def sort012(self, arr):
        zeros = 0
        ones = 0
        twos = 0
        for n in arr:
            if n == 0:
                zeros += 1
            elif n == 1:
                ones += 1
            else:
                twos += 1
        sorted_list = []
        idx = 0
    
        # place all the 0s
        for i in range(zeros):
            arr[idx] = 0
            idx += 1
    
        # place all the 1s
        for i in range(ones):
            arr[idx] = 1
            idx += 1
    
        # place all the 2s
        for i in range(twos):
            arr[idx] = 2
            idx += 1


```
